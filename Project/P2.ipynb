```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python

iris = pd.read_csv('iris.csv')
#print(iris.columns)
data = pd.DataFrame()
data[['Id','sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']] = iris[['Id', 'SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm',
       'Species']]
print(data.head())
```

```python
# create a figure and axis
fig, ax = plt.subplots()

# scatterplot
ax.scatter(data['sepal_width'], data['petal_length'])

# create labels and title

ax.set_title('Iris1')
ax.set_xlabel('Sepal WIdth')
ax.set_ylabel('Petal_length')
```

```python
# using colors for classes

# create a figure and axis
fig, ax = plt.subplots()

# create a color dictionary
color = {'Iris-setosa':'r', 'Iris-versicolor':'g', 'Iris-virginica':'b'}

# scatterplot
for i in range(len(data['sepal_length'])):
    ax.scatter(data['sepal_width'][i], data['petal_length'][i], color = color[data['class'][i]])

# create labels and title

ax.set_title('Iris1')
ax.set_xlabel('Sepal WIdth')
ax.set_ylabel('Petal_length')
```
