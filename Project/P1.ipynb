```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
titanic = sns.load_dataset("titanic")
print(titanic.head())
```

```python
# Set plot style
sns.set(style = "whitegrid") # try different styles: 'whitegrid', 'darkgrid', 'white', 'whitegrid'
```

```python
# Plot vertical bar chart with counts
#ax = sns.countplot(x = "class", data = titanic)
#ax.set(title = 'Titanic passengers by class', xlabel = 'Class', ylabel = 'Count')
ax = sns.countplot(x = "class", data = titanic).set(title = 'Titanic passengers by class', xlabel = 'Class', ylabel = 'Count')
ax
```

```python
# Plot horizontal bar chart with counts
#ax = sns.countplot(x = "class", data = titanic)
#ax.set(title = 'Titanic passengers by class', xlabel = 'Count', ylabel = 'Class')
ax = sns.countplot(y = "class", data = titanic).set(title = 'Titanic passengers by class', xlabel = 'Count', ylabel = 'Class')
ax
```

```python
# Plot vertical bar chart
ax = sns.barplot(x = "class", y = 'fare', data = titanic, ci = None)
#ax = sns.barplot(x = "class", y = 'fare', data = titanic, ci = "sd")
#ax = sns.barplot(x = "class", y = 'fare', data = titanic)
ax.set(title = 'Titanic fares by class', xlabel = 'Class', ylabel = 'Average Fare')
#ax = sns.barplot(x = "class", y = 'fare', data = titanic, ci = None).set(title = 'Titanic fares by class', xlabel = 'Class', ylabel = 'Average Fare')
```

```python
# Plot grouped count chart
ax = sns.countplot(x = "alive", hue = 'class', data = titanic)
ax.set(title = 'Titanic survivors and deaths by class', xlabel = 'Survived', ylabel = 'Count')
#ax = sns.countplot(x = "alive", hue = 'class', data = titanic).set(title = 'Titanic survivors and deaths by class', xlabel = 'Survived', ylabel = 'Count')
```

```python
# Plot grouped count chart
ax = sns.countplot(x = "class", hue = 'alive', data = titanic)
ax.set(title = 'Titanic survivors and deaths by class', xlabel = 'Class', ylabel = 'Count')
#ax = sns.countplot(x = "alive", hue = 'class', data = titanic).set(title = 'Titanic survivors and deaths by class', xlabel = 'Survived', ylabel = 'Count')
```

```python
# Plot grouped bar chart
ax = sns.barplot(x='class', y= 'fare', hue = 'alive', data = titanic, ci = None)
ax.set(title = 'Fare by class', xlabel = 'Class', ylabel = 'Fare')
```

```python
# Plot grouped bar chart
ax = sns.barplot(x='alive', y= 'fare', hue = 'class', data = titanic, ci = None)
ax.set(title = 'Fare by class', xlabel = 'Survived', ylabel = 'Fare')
```

```python
# Plot pie chart
a = titanic[titanic.pclass == 1]['pclass'].count()
b = titanic[titanic.pclass == 2]['pclass'].count()
c = titanic[titanic.pclass == 3]['pclass'].count()
plt.pie([a, b, c], labels = ['First', 'Second', 'Third'])
plt.title('Titanic passengers by class')
plt.axis('equal')
```

```python
# Plot scatter plot
ax = sns.regplot(titanic['age'], titanic['fare'])
ax.set(title = 'Titanic fares by age', xlabel = 'Age', ylabel = 'Fare')
#ax = sns.regplot(titanic['age'], titanic['fare']).set(title = 'Titanic fares by age', xlabel = 'Age', ylabel = 'Fare')
```

```python
# Plot scatter plot
ax = sns.relplot(x ='age', y = 'fare', data =titanic)
ax.set(title = 'Titanic fares by age', xlabel = 'Age', ylabel = 'Fare')
#ax = sns.regplot(titanic['age'], titanic['fare']).set(title = 'Titanic fares by age', xlabel = 'Age', ylabel = 'Fare')
```

```python
# Plot strip plot
ax = sns.stripplot(x = 'class', y = 'fare', data = titanic, jitter = False)
ax.set(title = 'Titanic fares by class', xlabel = 'Class', ylabel = 'Fare')
#ax = sns.stripplot(x = 'class', y = 'fare', data = titanic, jitter = False).set(title = 'Titanic fares by class', xlabel = 'Class', ylabel = 'Fare')
```

```python
# Plot strip plot with jittering
ax = sns.stripplot(x = 'class', y = 'fare', data = titanic, jitter = True)
ax.set(title = 'Titanic fares by class', xlabel = 'Class', ylabel = 'Fare')
```

```python
# Plot swarm plot
ax = sns.swarmplot(x = 'class', y = 'fare', data = titanic)
ax.set(title = 'Titanic fares by class', xlabel = 'Class', ylabel = 'Fare')
```

```python
# Plot histogram
heights = pd.read_csv('data_lecture_06_b.csv')
plt.title('Height histogram')
plt.xlabel('Height')
plt.ylabel('Frequency')
plt.hist(heights['Height'], bins = 6)
```

```python
# Plot box plot
tips = sns.load_dataset('tips')
print(tips)
ax = sns.boxplot(x = 'time', y = 'total_bill', data = tips)
ax.set(title = 'Box plot of total bill by time of day', xlabel = 'Time of day', ylabel = 'Total Bill')
#ax = sns.boxplot(x = 'time', y = 'total_bill', data = tips).set(title = 'Box plot of total bill by time of day', xlabel = 'Time of day', ylabel = 'Total Bill')
```

```python
# Plot box plot without outliers
ax = sns.boxplot(x = 'time', y = 'total_bill', data = tips, whis = 10)
ax.set(title = 'Box plot of total bill by time of day', xlabel = 'Time of day', ylabel = 'Total Bill')
#ax = sns.boxplot(x = 'time', y = 'total_bill', data = tips, whis = 10).set(title = 'Box plot of total bill by time of day', xlabel = 'Time of day', ylabel = 'Total Bill')
```

```python
# Plot violin plot
#Violin plot shows the denstiy of distribution
ax = sns.violinplot(x = 'time', y = 'total_bill', data = tips)
ax.set(title = 'Violin plot of total bill by time of day', xlabel = 'Time of day', ylabel = 'Total Bill')
#ax = sns.violinplot(x = 'time', y = 'total_bill', data = tips).set(title = 'Violin plot of total bill by time of day', xlabel = 'Time of day', ylabel = 'Total Bill')
```
